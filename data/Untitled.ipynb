```python
from datetime import datetime

# Current year to calculate age
current_year = datetime.now().year

for n in range(len(data["YOB"])):
    yob = data["YOB"][n]
    # Check if the value is a date format (contains '/')
    if '/' in yob:
        # Extract the year part from the full date (e.g. '20/05/1997' -> '1997')
        yob = yob.split('/')[-1]
    # Handle 4-digit year
    if len(yob) > 4:
        # If it’s already a 4-digit year, use it
        year_of_birth = int(yob)
    # Handle 2-digit year
    elif len(yob) == 2:
        # Assuming the 2-digit year is in the 2000s (adjust logic if needed)
        year_of_birth = current_year - (current_year % 100) + int(yob)
    # Handle the case of 4-digit year already formatted correctly
    elif len(yob) == 4:
        year_of_birth = int(yob)
    # Store the corrected year back into the data dictionary
    data["YOB"][n] = str(year_of_birth)
# Calculate the ages and print
for yob in data["YOB"]:
    print(f"YOB: {yob}, Age: {current_year - int(yob)}")
```

```python
data['YOB'] = pandas.to_numeric(data['YOB'], errors='coerce')  # Ensure YOB is numeric, replace non-numeric with NaN
data.loc[(data['YOB'] < 1920) | (data['YOB'] > current_year), 'YOB'] = None

# Convert YOB to Age
data['Age'] = current_year - data['YOB']

# Drop rows with invalid ages (optional)
data_cleaned = data.dropna(subset=['Age'])

# Show the first few rows of the cleaned data
data_cleaned.head()
```
